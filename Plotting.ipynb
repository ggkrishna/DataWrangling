# Plotting in Python

### Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,2*np.pi,100)
plt.plot(np.cos(x)**2) # just a single array 
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
y=np.cos(x)**2
plt.plot(x,y) # array of points(x,y)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
y1=np.cos(x)**2
y2=np.sin(x)**2
plt.plot(x,y1, x,y2) # two arrays of points
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
y1=np.cos(x)**2
y2=np.sin(x)**2
plt.plot(x,y1,'g', x,y2,'b') # two arrays of points seperated by format string
plt.show()
```

```python
# More than two arrays
x=np.linspace(0,2*np.pi,100)
y1=np.cos(x)**2
y2=np.sin(x)**2
y3=np.ones_like(y1)*0.5
plt.plot(x,y1,'g', x,y2,'b',x,y3,'r') 
plt.show()
```

```python
# other parameters
x=np.linspace(0,2*np.pi,100)
y1=np.cos(x)**2
y2=np.sin(x)**2
plt.plot(x,y1,'g.--', x,y2,'b.--',linewidth=2, markersize=2) # two arrays of points
plt.show()
```

```python
# Better flexibility on each function
x=np.linspace(0,2*np.pi,100)
y1=np.cos(x)**2
y2=np.sin(x)**2
plt.plot(x,y1,'g.--', linewidth=3, markersize=12,)  
plt.plot(x,y2,'b.--', linewidth=1,markersize=4) 
plt.show()
```

```python
# format string not as shortcut
x=np.linspace(0,2*np.pi,100)
y1=np.cos(x)**2
plt.plot(x, y1,color='green', marker=',', linestyle='dashed',linewidth=2, markersize=12,) 
plt.show()
```

```python
# labels
x=np.linspace(0,2*np.pi,100)
y1=np.cos(x)**2
y2=np.sin(x)**2
plt.plot(x,y1,'g.--', linewidth=3, markersize=12,label="grees")  
plt.plot(x,y2,'b.--', linewidth=1,markersize=4,label="blues")
plt.legend(loc=0)
plt.show()
```

```python
# subplots
```

```python
f,a=plt.subplots(1,1)
a.set(title='My title', ylabel='Y-Axis', xlabel='X-Axis')
ln,=a.plot(x,x)
plt.show()
```

```python
f,a=plt.subplots(1,1)
a.set(title='My title', ylabel='Y-Axis', xlabel='X-Axis')
x=np.linspace(0,2*np.pi,100)
p1,=a.plot(x,x)
p1.set(label="straight",linewidth=2)
p2,=a.plot(x,np.sin(x))
p2.set(label="sin")
plt.legend()
plt.show()
```

```python
f,a=plt.subplots(1,2,sharey=False)
a[0].set(title='First plot', ylabel='Y1', xlabel='X1')
a[1].set(title='Second plot', ylabel='Y2', xlabel='X2')
x=np.linspace(0,2*np.pi,100)
a[0].plot(x,x)
a[1].plot(x,np.sin(x))
a[1].plot(x,50+np.sin(x))
f.tight_layout()
plt.show()
```

```python
f,(ln,sn)=plt.subplots(1,2,sharey=False)
ln.set(title='$\omega$', ylabel='Y1', xlabel='X1')
sn.set(title='Second plot', ylabel='Y2', xlabel='X2')
x=np.linspace(0,2*np.pi,100)
ln.plot(x,x)
sn.plot(x,np.sin(x))
sn.plot(x,50+np.sin(x))
f.tight_layout()
plt.show()
```

```python
f,a=plt.subplots(2,1,sharey=False)
a[0].set(title='First plot', ylabel='Y1', xlabel='X1')
a[1].set(title='Second plot', ylabel='Y2', xlabel='X2')
x=np.linspace(0,2*np.pi,100)
a[0].plot(x,x)
a[1].plot(x,np.sin(x))
a[1].plot(x,50+np.sin(x))
plt.suptitle("vertical plots")
f.tight_layout()
plt.show()
```
